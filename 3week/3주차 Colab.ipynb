```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow import keras

df = pd.read_csv("/content/drive/MyDrive/AI_class/car_evaluation.csv", header=None)
df
```

```python
df.columns = ["buying", "maint", "doors", "persons", "lug_capacity", "safety", "output"]
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 인코딩 (숫자로 변환)
label_encoders = {}
for col in df.columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  # 데이터를 변환하여 저장
    label_encoders[col] = le  # 나중에 역변환할 수 있도록 저장

# 데이터 확인
df
```

```python
# 레이블링
df['output'].value_counts()
```

```python
# 데이터와 타겟 분리
X = df.drop('output', axis=1).values
y = df['output'].values
```

```python
X
```

```python
y
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 훈련, 테스트 데이터로 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Build a logistic regression classifier and predict

clf_lr = LogisticRegression(random_state=0)
clf_lr.fit(X_train, y_train)

pred_lr = clf_lr.predict(X_test)

print ("\n--- Logistic Regression Classifier ---")
print (accuracy_score(y_test, pred_lr))
print (confusion_matrix(y_test, pred_lr))
```

```python
# Build a decision tree classifier and predict

clf_dt = DecisionTreeClassifier(random_state=0)
clf_dt.fit(X_train, y_train)

pred_dt = clf_dt.predict(X_test)

print ("\n--- Decision Tree Classifier ---")
print (accuracy_score(y_test, pred_dt))
print (confusion_matrix(y_test, pred_dt))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

print ("\n--- Radom Forest ---")
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
print(accuracy_score(y_test,pred))
print (confusion_matrix(y_test, pred))
```

```python
from sklearn.svm import SVC

clf_svm = SVC(random_state=0)
clf_svm.fit(X_train, y_train)

pred_svm = clf_svm.predict(X_test)

print("\n--- SVM Classifier ---")
print(accuracy_score(y_test, pred_svm))
print(confusion_matrix(y_test, pred_svm))
```

```python
df = pd.read_csv("/content/drive/MyDrive/AI_class/diabetes.csv")
df
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# Outcome 칼럼 삭제
df = df.drop(columns=['Outcome'])
df
```

```python
# 레이블링
df['BMI'].value_counts()
```

```python
# 데이터와 타겟 분리
X = df.drop('BMI', axis=1).values
y = df['BMI'].values
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,
                                                  y,
                                                  test_size=0.2,
                                                  shuffle=True,
                                                  random_state=12)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## RandomForestRegressor

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

## SVR

```python
model = SVR()
model.fit(X_train, y_train)
dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

## DecisionTreeRegressor

```python
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```

## LinearRegression

```python
model = LinearRegression()
model.fit(X_train, y_train)
dy_preds = model.predict(X_test)
print('평균제곱근오차', mean_squared_error(dy_preds, y_test))
```
